## 성능 비교 (yolov8n.pt vs best.pt)

- Precision: 모델이 "객체다"라고 예측했을 때, 실제 맞을 확률
- Recall: 실제 객체를 놓치지 않고 찾아낸 비율이 93.1%
- mAP@0.5: 모델이 얼마나 정확하게 객체를 찾아내고, 그 위치까지 잘 맞췄는지를 평균적으로 나타냄.


| 지표                 | yolov8n.pt | best.pt | 변화   |
|----------------------|-----------------|-------------------|--------|
| Precision (정밀도)   | 0.922           | 0.955             | 🔼 상승 |
| Recall (재현율)      | 0.855           | 0.918             | 🔼 상승 |
| mAP@0.5              | 0.916           | 0.950             | 🔼 상승 |
| mAP@0.5:0.95         | 0.691           | 0.759             | 🔼 상승 |
| Fitness (종합지표)   | 0.713           | 0.778             | 🔼 상승 |


## best_parameters.yaml

```python
from ultralytics import YOLO
# YOLOv8 모델 로드
model = YOLO("yolov8n.pt")
# 하이퍼파라미터 탐색 수행
model.tune(
data="./datasets/data.yaml",
epochs=5,
imgsz=640,
iterations=5, # 탐색 횟수 (예: 25번의 실험)
val=True
)
```

```python
from ultralytics import YOLO
import yaml

model = YOLO("yolov8n.pt")
data_path = "./datasets/data.yaml"

# 튜닝 결과 하이퍼파라미터 파일 불러오기
with open("runs/detect/tune2/best_hyperparameters.yaml", "r", encoding="utf-8") as f:
    hyp_dict = yaml.safe_load(f)

# 최적 하이퍼파라미터로 학습
model.train(
    data=data_path,
    epochs=20,   # 원하는 만큼 학습 (예: 20, 50, 100)
    imgsz=640,
    batch=32,
    **hyp_dict
)
```
